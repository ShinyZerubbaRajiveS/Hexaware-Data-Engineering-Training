##**Section 1 — Basic Queries**


```sql
-- 1. Show all employees
SELECT * FROM employees;

```

```sql
-- 2. Show only emp_name , city , salary
SELECT emp_name, city, salary FROM employees;

```

```sql
-- 3. Show distinct departments
SELECT DISTINCT department FROM employees;

```

```sql
-- 4. Count total employees
SELECT COUNT(*) FROM employees;

```

```sql
-- 5. Show employees from Hyderabad
SELECT * FROM employees WHERE city = 'Hyderabad';

```

```sql
-- 6. Show employees from Bangalore
SELECT * FROM employees WHERE city = 'Bangalore';

```

```sql
-- 7. Show employees whose salary is greater than 90000
SELECT * FROM employees WHERE salary > 90000;

```

```sql
-- 8. Show employees whose experience is greater than 6 years
SELECT * FROM employees WHERE experience > 6;

```

```sql
-- 9. Show employees working in Engineering department
SELECT * FROM employees WHERE department = 'Engineering';

```

```sql
-- 10. Show employees whose salary is between 80000 and 95000
SELECT * FROM employees WHERE salary BETWEEN 80000 AND 95000;
```

```sql
-- 16. Find average salary of all employees
SELECT AVG(salary) FROM employees;
```

```sql
-- 19. Count employees per department
SELECT department, COUNT(*) FROM employees GROUP BY department;
```

##**Section 5 — Conditional Queries**


```sql
-- 26. Show employees with salary greater than company average
SELECT * FROM employees 
WHERE salary > (SELECT AVG(salary) FROM employees);
```

```sql
-- 27. Show employees with experience greater than department average
SELECT * FROM employees e
WHERE experience > (
    SELECT AVG(experience) 
    FROM employees 
    WHERE department = e.department
);
```

```sql
-- 28. Show employees whose salary is less than 70000
SELECT * FROM employees WHERE salary < 70000;
```

```sql
-- 29. Show employees working in HR department
SELECT * FROM employees WHERE department = 'HR';
```

```sql
-- 30. Show employees working in Finance department with salary greater than 90000
SELECT * FROM employees 
WHERE department = 'Finance' AND salary > 90000;
```

##**Section 6 — Joins**


```sql
-- Show employees with project budgets by joining employees and projects tables using department
SELECT e.*, p.budget
FROM employees e
JOIN projects p
ON e.department = p.department;
```

```sql
-- Show employee name, department, project name
SELECT e.emp_name, e.department, p.project_name
FROM employees e
JOIN projects p
ON e.department = p.department;
```

```sql
-- 33. Show department-wise project budget
SELECT p.department, SUM(p.budget)
FROM projects p
GROUP BY p.department;
```

```sql
-- 34. Show employees working in departments that have projects
SELECT DISTINCT e.*
FROM employees e
JOIN projects p
ON e.department = p.department;
```

```sql
-- 35. Count number of employees working per project
SELECT p.project_name, COUNT(e.emp_id)
FROM projects p
JOIN employees e
ON p.department = e.department
GROUP BY p.project_name;
```

##**Section 7 — Advanced Queries**


```sql
-- 36. Find top 3 highest paid employees in Engineering department
SELECT * FROM employees 
WHERE department = 'Engineering'
ORDER BY salary DESC 
LIMIT 3;
```

```sql
-- 37. Find department with highest average salary
SELECT department
FROM employees
GROUP BY department
ORDER BY AVG(salary) DESC
LIMIT 1;
```

```sql
-- 38. Find city with highest number of employees
SELECT city
FROM employees
GROUP BY city
ORDER BY COUNT(*) DESC
LIMIT 1;
```

```sql
-- 39. Find employees whose salary is higher than department average
SELECT * FROM employees e
WHERE salary > (
    SELECT AVG(salary)
    FROM employees
    WHERE department = e.department
);
```

```sql
-- 40. Find employees whose experience is highest in their department
SELECT * FROM employees e
WHERE experience = (
    SELECT MAX(experience)
    FROM employees
    WHERE department = e.department
);
```

##Final Mini Capstone

Write SQL queries to create an analytics table containing: emp_name city department salary experience salary_category department_avg_salary 
Rules: 
salary_category salary >= 95000 → High 
salary >= 80000 → Medium 
else → Low 


```sql
CREATE TABLE employee_salary_analysis AS
SELECT 
    emp_name,
    city,
    department,
    salary,
    experience,
    CASE 
        WHEN salary >= 95000 THEN 'High'
        WHEN salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,
    AVG(salary) OVER (PARTITION BY department) AS department_avg_salary
FROM employees;
```

```sql
SELECT * FROM employee_salary_analysis;
```
