```python
from pyspark.sql.functions import col, sum, count 
data = [ (1, "Aarav Sharma", "Hyderabad", "Laptop", 1, 75000), (2, "Priya Reddy", "Bengaluru", "Mouse", 4, 600), (3, "Rohan Mehta", "Mumbai", "Monitor", 2, 12000), (4, "Sneha Iyer", "Chennai", "Keyboard", 3, 1500), (5, "Kiran Patel", "Hyderabad", "Laptop", 1, 75000), (6, "Ananya Das", "Kolkata", "Tablet", 2, 30000), (7, "Vikram Singh", "Delhi", "Mouse", 6, 600), (8, "Meera Nair", "Bengaluru", "Monitor", 1, 12000), (9, "Rahul Verma", "Mumbai", "Laptop", 1, 75000), (10, "Pooja Shah", "Hyderabad", "Keyboard", 5, 1500) ] 

columns = ["sale_id", "customer_name", "city", "product", "quantity", "unit_price"] 
df = spark.createDataFrame(data, columns) 
display(df) 
```

```python
# 1. Add revenue column
df = df.withColumn("revenue", col("quantity") * col("unit_price"))
display(df)

# 2. Total revenue per city
revenue_per_city = df.groupBy("city") \
    .agg(sum("revenue").alias("total_revenue"))
display(revenue_per_city)

# 3. Total quantity sold per city
quantity_per_city = df.groupBy("city") \
    .agg(sum("quantity").alias("total_quantity"))
display(quantity_per_city)

# 4. Sort cities by total revenue (descending)
revenue_sorted = df.groupBy("city") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())
display(revenue_sorted)
```
