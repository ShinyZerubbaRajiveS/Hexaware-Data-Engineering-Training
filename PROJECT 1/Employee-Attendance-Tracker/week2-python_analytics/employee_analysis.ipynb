```python
import pandas as pd
import numpy as np
```

```python
employees = pd.read_csv("employees.csv")
attendance = pd.read_csv("attendance.csv")
tasks = pd.read_csv("tasks.csv")
```

```python
# =========================================
# CLEAN DATA
# =========================================

attendance.fillna("Unknown", inplace=True)
tasks.fillna(0, inplace=True)

# Remove duplicate rows if any
attendance.drop_duplicates(inplace=True)
tasks.drop_duplicates(inplace=True)

print("Datasets cleaned successfully!")

# =========================================
# MERGE DATASETS
# =========================================

merged_data = pd.merge(employees, attendance, on="employee_id")
merged_data = pd.merge(merged_data, tasks, on="employee_id")

# =========================================
# CALCULATE WORK HOURS
# =========================================

merged_data["clock_in"] = pd.to_datetime(
    merged_data["clock_in"],
    format="%H:%M"
)

merged_data["clock_out"] = pd.to_datetime(
    merged_data["clock_out"],
    format="%H:%M"
)

merged_data["work_hours"] = (
    merged_data["clock_out"] -
    merged_data["clock_in"]
).dt.total_seconds() / 3600

# =========================================
# PRODUCTIVITY SCORE USING NUMPY
# =========================================

merged_data["productivity_score"] = np.where(
    merged_data["status"] == "Absent",
    0,
    merged_data["tasks_completed"] / merged_data["work_hours"]
)

# =========================================
# TOP PERFORMERS
# =========================================

top_performers = merged_data.sort_values(
    by="productivity_score",
    ascending=False
)

print("\nTop Performers")
print(top_performers[[
    "employee_name",
    "productivity_score"
]].head(5))

# =========================================
# BOTTOM PERFORMERS
# =========================================

bottom_performers = merged_data.sort_values(
    by="productivity_score"
)

print("\nBottom Performers")
print(bottom_performers[[
    "employee_name",
    "productivity_score"
]].head(5))

# =========================================
# FREQUENT ABSENTEES
# =========================================

absentees = merged_data[
    merged_data["status"] == "Absent"
]

print("\nFrequent Absentees")
print(absentees[[
    "employee_name",
    "department",
    "status"
]])

# =========================================
# SAVE REPORT
# =========================================

merged_data.to_csv(
    "employee_productivity_report.csv",
    index=False
)

print("\nReport Generated Successfully!")
```
