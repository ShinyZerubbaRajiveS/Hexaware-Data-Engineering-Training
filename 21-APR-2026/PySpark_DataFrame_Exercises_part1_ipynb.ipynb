###**PART 1 — DataFrame Creation**

**Exercise 1**

Create a Spark session.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('Datastructure examples').getOrCreate()

print("Spark session created...!")
```

**Exercise 2**

Load employees.csv .

```python
employees_df=spark.read.csv("employees.csv",header=True,inferSchema=True)
employees_df.show()
```

**Exercise 3**

Load departments.csv .

```python
departments_df=spark.read.csv("departments.csv",header=True,inferSchema=True)
departments_df.show()
```

**Exercise 4**

Load sales.csv .

```python
sales_df=spark.read.csv("sales.csv",header=True,inferSchema=True)
sales_df.show()
```

**Exercise 5**

Load employees_nested.json .

```python
emp_nested_df = spark.read.option("multiline", "true").json("employees_nested.json")
emp_nested_df.show(truncate=False)
```

**Exercise 6**

Print schema for all DataFrames.

```python
print("Employees Schema:")
employees_df.printSchema()

print("Departments Schema:")
departments_df.printSchema()

print("Sales Schema:")
sales_df.printSchema()

print("Nested JSON Schema:")
emp_nested_df.printSchema()
```

###**PART 3 — Selecting and Renaming**

**Exercise 11**

Select only:
name
salary

```python
employees_df.select("name", "salary").show()
```

**Exercise 12**

Rename column salary to emp_salary .

```python
emp_df_renamed = employees_df.withColumnRenamed("salary", "emp_salary")
emp_df_renamed.show()
```

**Exercise 13**

Select employees who joined after 2023-01-01.

```python
from pyspark.sql.functions import col, to_date
employees_df.filter(
    to_date(col("joining_date"), "yyyy-MM-dd") > "2023-01-01"
).show()
```

###**PART 4 — Filtering Data**

**Exercise 14**

Find employees with salary greater than 65000.

```python
employees_df.filter(col("salary") > 65000).show()
```

**Exercise 15**

Find employees in IT department.

```python
employees_df.filter(col("department") == "IT").show()
```

**Exercise 16**

Find employees in IT department with salary greater than 70000.

```python
employees_df.filter((col("department") == "IT") &
 (col("salary") > 70000)).show()
```

###**PART 5 — Dropping and Sorting**

**Exercise 17**

Drop column joining_date .

```python
employees_df_dropped = employees_df.drop("joining_date")
employees_df_dropped.show()
```

**Exercise 18**

Sort employees by salary ascending.

```python
employees_df.orderBy("salary").show()
```

**Exercise 19**

Sort employees by salary descending.

```python
employees_df.orderBy(col("salary").desc()).show()
```

**Exercise 20**

Find top 3 highest paid employees.

```python
employees_df.orderBy(col("salary").desc()).show(3)
```

###**PART 6 — Aggregations**

**Exercise 21**

Find total salary of all employees.

```python
from pyspark.sql.functions import sum
employees_df.select(sum("salary").alias("total_salary")).show()
```

**Exercise 22**

Find average salary.

```python
from pyspark.sql.functions import avg
employees_df.select(avg("salary").alias("avg_salary")).show()
```

**Exercise 23**

Find highest salary.

```python
from pyspark.sql.functions import max
employees_df.select(max("salary").alias("highest_salary")).show()
```

**Exercise 24**

Find lowest salary.

```python
from pyspark.sql.functions import min
employees_df.select(min("salary").alias("lowest_salary")).show()
```

###**PART 7 — GroupBy**

**Exercise 25**

Count employees per department.

```python
employees_df.groupBy("department").count().show()
```

**Exercise 26**

Calculate average salary per department.

```python
from pyspark.sql.functions import avg
employees_df.groupBy("department").agg(avg("salary").
                                       alias("avg_salary")).show()
```

**Exercise 27**

Find total salary per department.

```python
from pyspark.sql.functions import sum
employees_df.groupBy("department").agg(sum("salary").
                                    alias("total_salary")).show()
```

###**PART 8 — Joining DataFrames**
Use employees.csv and departments.csv .

**Exercise 28**

Perform inner join.

```python
inner_df=employees_df.join(departments_df, on="department", how="inner")
inner_df.show()
```

**Exercise 29**

Perform left join.

```python
left_df=employees_df.join(departments_df, on="department", how="left")
left_df.show()
```

**Exercise 30**

Show employee name with department location.

```python
employees_df.join(departments_df, on="department", how="left").select("name", "department","location").show()
```

###**PART 9 — Updating Data**

**Exercise 31**

Add a column bonus = salary * 10%.

```python
employees_df = employees_df.withColumn("bonus",
                                       col("salary") * 0.10)
employees_df.show()
```

**Exercise 32**

Add a column company with value "BotCampus" .

```python
from pyspark.sql.functions import lit

employees_df = employees_df.withColumn("company",
    lit("BotCampus")
)
employees_df.show()
```

###**PART 10 — Handling Null Values**
Note employee Meera has null department.

**Exercise 33**

Find rows with null department.

```python
employees_df.filter(col("department").isNull()).show()
```

**Exercise 34**

Replace null department with "Unknown" .

```python
employees_df_filled = employees_df.fillna({
    "department": "Unknown"})
employees_df_filled.show()
```

**Exercise 35**

Drop rows with null values.

```python
employees_df_dropped = employees_df.dropna()
employees_df_dropped.show()
```

###**PART 11 — Built-in Functions**

**Exercise 36**

Convert employee names to uppercase.

```python
from pyspark.sql.functions import upper

employees_df_upper = employees_df.withColumn("name_upper",
                                             upper(col("name")))
employees_df_upper.show()
```

**Exercise 37**

Find length of employee names.

```python
from pyspark.sql.functions import length

employees_df_length = employees_df.withColumn("name_length",
                                              length(col("name")))
employees_df_length.show()
```

**Exercise 38**

Create column salary_in_lakhs.

```python
employees_df_lakhs = employees_df.withColumn("salary_in_lakhs",
                                        col("salary") / 100000)
employees_df_lakhs.show()
```
