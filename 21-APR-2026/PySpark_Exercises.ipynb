```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('PySpark Exercises').getOrCreate()

print("Spark session created...!")
```


### Exercise 1 Create a DataFrame with:
- Name
Salary
Department

- Add 5 employees.



```python
data=[
    ('James',60000,"HR"),
    ('Michael',45000,"IT"),
    ('Robert',70000,"Finance"),
    ('Maria',52000,"IT"),
    ('Alice',48000,"HR")
]
columns=["Name", "Salary", "Department"]
df=spark.createDataFrame(data, columns)
df.show()
```

### Exercise 2 Show only Name and Salary columns.

```python
df.select("Name", "Salary").show()
```

### Exercise 3 Filter employees with salary greater than 50000.

```python
df.filter(df.Salary > 50000).show()
```

### Exercise 4 Count employees per department.

```python
df.groupBy("Department").count().show()
```

### Exercise 5 Sort employees by salary.

```python
df.orderBy("Salary").show()
```
