```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('RDD Exercises').getOrCreate()
sc= spark.sparkContext
print("Spark session created...!")
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]
logs_rdd = sc.parallelize(logs)
logs_rdd.collect()
```

```python
sales = [
("Laptop",1,75000),
("Mouse",3,500),
("Keyboard",2,1500),
("Laptop",1,75000),
("Monitor",1,12000),
("Mouse",2,500),
("Keyboard",1,1500),
("Laptop",1,75000),
("Monitor",2,12000),
("Mouse",3,500),
("Laptop",1,75000),
("Keyboard",2,1500),
("Monitor",1,12000),
("Mouse",1,500),
("Laptop",1,75000)
]
sales_rdd = sc.parallelize(sales)
sales_rdd.collect()
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),

("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

##**Part 1 — Basic RDD Exercises**

**Exercise 1**

Count total log entries.
Expected function:
count()

```python
logs_rdd.count()
```

**Exercise 2**

Print first 5 log records.

```python
logs_rdd.take(5)
```

**Exercise 3**

Extract only user names.
Expected output:
["Rahul","Sneha","Arjun"...]
Hint:
split()
map()

```python
users = logs_rdd.map(lambda x: x.split()[0])
users.collect()
```

**Exercise 4**

Extract only actions.
Expected output:
["login","purchase","logout"]

```python
actions = logs_rdd.map(lambda x: x.split()[1])
actions.collect()
```

**Exercise 5**

Find unique users.
Expected concept:
distinct()

```python
unique_users = users.distinct()
unique_users.collect()
```

**Exercise 6**

Count number of login actions.

```python
login_count = logs_rdd.filter(lambda x: x.split()[1] == "login").count()
login_count
```

**Exercise 7**

Find all purchase actions.

```python
purchases = logs_rdd.filter(lambda x: x.split()[1] == "purchase")
purchases.collect()
```

##**Part 2 — flatMap Practice**

```python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
```

**Exercise 8**

Split all sentences into words.
Expected concept:
flatMap

```python
words = sentences.flatMap(lambda x: x.split())
words.collect()
```

**Exercise 9**

Find all distinct technologies.
Expected output example:
["spark","hadoop","hive","python","kafka"]

```python
distinct_words = words.distinct()
distinct_words.collect()
```

**Exercise 10**

Count number of occurrences of each word.
Expected result format:
spark → 4
hadoop → 2
python → 1

Hint pipeline:
flatMap
map
reduceByKey

```python
word_count = words.map(lambda x: (x, 1)) \
                  .reduceByKey(lambda a, b: a + b)

word_count.collect()
```

##**Part 3 — Sales Dataset Exercises**

**Exercise 11**

Calculate revenue per row.
Formula:
quantity × price

Expected format:
(product, revenue)

```python
revenue_rdd = sales_rdd.map(lambda x: (x[0], x[1] * x[2]))
revenue_rdd.collect()
```

**Exercise 12**

Calculate total revenue of all sales.
Expected concept:
reduce()

```python
total_revenue = revenue_rdd.map(lambda x: x[1]) \
                           .reduce(lambda a, b: a + b)

total_revenue
```

**Exercise 13**

Find distinct products sold.

```python
products = sales_rdd.map(lambda x: x[0]).distinct()
products.collect()
```

**Exercise 14**

Find all products with quantity greater than 1.

```python
high_qty = sales_rdd.filter(lambda x: x[1] > 1)
high_qty.collect()
```

**Exercise 15**

Calculate total quantity sold per product.
Expected format:
Laptop → 5
Mouse → 9
Keyboard → 5

Hint:
map → (product, quantity)
reduceByKey

```python
qty_per_product = sales_rdd.map(lambda x: (x[0], x[1])) \
                           .reduceByKey(lambda a, b: a + b)

qty_per_product.collect()
```

**Exercise 16**

Calculate total revenue per product.
Expected output example:
Laptop → 375000
Mouse → 4500
Keyboard → 7500
Monitor → 48000

```python
rev_per_product = sales_rdd.map(lambda x: (x[0], x[1] * x[2])) \
                           .reduceByKey(lambda a, b: a + b)

rev_per_product.collect()
```

**Exercise 17**

Find the highest revenue product.

Hint:
sortBy
take

```python
top_product = rev_per_product.sortBy(lambda x: x[1], ascending=False) \
                            .take(1)

top_product
```

##**Part 4 — Employee Dataset**

**Exercise 18**

Extract only employee names.

```python
names = emp_rdd.map(lambda x: x[0])
names.collect()
```

**Exercise 19**

Find employees with salary greater than 70,000.

```python
high_salary = emp_rdd.filter(lambda x: x[2] > 70000)
high_salary.collect()
```

**Exercise 20**

Find distinct departments.

```python
departments = emp_rdd.map(lambda x: x[1]).distinct()
departments.collect()
```

**Exercise 21**

Calculate total salary per department.
Expected output:
IT → total salary
HR → total salary
Finance → total salary

```python
total_salary_dept = emp_rdd.map(lambda x: (x[1], x[2])) \
                           .reduceByKey(lambda a, b: a + b)

total_salary_dept.collect()
```

**Exercise 22**

Calculate average salary per department.

Hint:
You will need:
map
reduceByKey

```python
avg_salary_dept = emp_rdd.map(lambda x: (x[1], (x[2], 1))) \
                         .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1])) \
                         .mapValues(lambda x: x[0] / x[1])

avg_salary_dept.collect()
```

**Exercise 23**

Find the highest paid employee.

```python
top_employee = emp_rdd.sortBy(lambda x: x[2], ascending=False) \
                      .take(1)

top_employee
```

##**Part 5 — Intermediate RDD Exercises**

**Exercise 24**

Count number of employees per department.
Expected output:
IT → 4
HR → 3
Finance → 3

```python
emp_count_dept = emp_rdd.map(lambda x: (x[1], 1)) \
                        .reduceByKey(lambda a, b: a + b)

emp_count_dept.collect()
```

**Exercise 25**

Sort employees by salary descending.

```python
sorted_emp = emp_rdd.sortBy(lambda x: x[2], ascending=False)
sorted_emp.collect()
```

**Exercise 26**

Find top 3 highest salaries.

```python
top3 = emp_rdd.takeOrdered(3, key=lambda x: -x[2])
top3
```

**Exercise 27**

Create (department, employee_name) pairs.

```python
dept_name_pairs = emp_rdd.map(lambda x: (x[1], x[0]))
dept_name_pairs.collect()
```

**Exercise 28**

Group employees by department.
Hint:
groupByKey

```python
grouped = dept_name_pairs.groupByKey()
result = grouped.mapValues(list)
result.collect()
```

## **Part 6 — Advanced RDD Thinking**

**Exercise 29**

Find department with highest total salary.

```python
total_salary_dept = emp_rdd.map(lambda x: (x[1], x[2])) \
                           .reduceByKey(lambda a, b: a + b)
top_dept = total_salary_dept.sortBy(lambda x: x[1], ascending=False) \
                            .take(1)
top_dept
```

**Exercise 30**

Find employee count and total salary per department.

Expected output format:
IT → (count, total_salary)

```python
dept_stats = emp_rdd.map(lambda x: (x[1], (1, x[2]))) \
                    .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))

dept_stats.collect()
```

## **Final Mini Project (RDD)**

Tasks:
1. Calculate total revenue per product
2. Find top selling product
3. Find product generating highest revenue
4. Find products sold more than 3 times
5. Sort products by revenue

Final expected output example:
Product → Revenue
Laptop → 375000
Monitor → 48000
Keyboard → 7500

```python
revenue_per_product = sales_rdd.map(lambda x: (x[0], x[1] * x[2])) \
                               .reduceByKey(lambda a, b: a + b)
revenue_per_product.collect()
```

```python
qty_per_product = sales_rdd.map(lambda x: (x[0], x[1])) \
                           .reduceByKey(lambda a, b: a + b)
top_selling = qty_per_product.sortBy(lambda x: x[1], ascending=False) \
                             .take(1)
top_selling
```

```python
top_revenue_product = revenue_per_product.sortBy(lambda x: x[1], ascending=False) \
                                         .take(1)
top_revenue_product
```

```python
more_than_3 = qty_per_product.filter(lambda x: x[1] > 3)
more_than_3.collect()
```

```python
sorted_revenue = revenue_per_product.sortBy(lambda x: x[1], ascending=False)
sorted_revenue.collect()
```
