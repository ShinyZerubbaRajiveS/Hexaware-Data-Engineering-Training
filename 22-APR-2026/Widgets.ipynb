```python
import ipywidgets as widgets
from IPython.display import display
```

##Text Widget

```python
name_widget=widgets.Text(
    value="Zerubba",
    description="Name:"
    )
display(name_widget)
```

##DropDown

```python
dept_widget=widgets.Dropdown(
    options=["IT","HR","Finance"],
    value="IT",
    description="Department:"
)
display(dept_widget)
```

##Slider

```python
sal_slider=widgets.IntSlider(
    value=60000,
    min=30000,
    max=100000,
    step=5000,
    description="Salary:"
)
display(sal_slider)
```

##Button

```python
button=widgets.Button(
    description="Show Values"
)
def on_button_clicked(b):
    print(f"Name:{name_widget.value}")
    print(f"Department:{dept_widget.value}")
    print(f"Salary:{sal_slider.value}")
button.on_click(on_button_clicked)
display(button)
```
