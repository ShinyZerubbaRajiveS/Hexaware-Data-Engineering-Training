##Part 2 — DataFrame Transformations

```python
from pyspark.sql.functions import col, to_date, when

# 1. Convert registration_date into date type
patients_df = patients_df.withColumn(
    "registration_date",
    to_date(col("registration_date"), "yyyy-MM-dd")
)

# 2. Convert visit_date into date type
visits_df = visits_df.withColumn(
    "visit_date",
    to_date(col("visit_date"), "yyyy-MM-dd")
)

# 3. Add a column test_cost (assume based on visit_status)
visits_df = visits_df.withColumn(
    "test_cost",
    when(col("visit_status") == "Completed", 500)
    .otherwise(200)
)

# 4. Add a column estimated_total_bill (consultation + test)
full_df = visits_df.join(doctors_df, "doctor_id", "left")

full_df = full_df.withColumn(
    "estimated_total_bill",
    col("consultation_fee") + col("test_cost")
)

# 5. Create age_category
patients_df = patients_df.withColumn(
    "age_category",
    when(col("age") < 18, "Child")
    .when((col("age") >= 18) & (col("age") <= 60), "Adult")
    .otherwise("Senior")
)

# 6. Create membership_priority
patients_df = patients_df.withColumn(
    "membership_priority",
    when(col("membership") == "Gold", 1)
    .when(col("membership") == "Silver", 2)
    .otherwise(3)
)

# 7. Create visit_priority
visits_df = visits_df.withColumn(
    "visit_priority",
    when(col("visit_status") == "Completed", "High")
    .otherwise("Low")
)

# 8. Create high_bill_flag
full_df = full_df.withColumn(
    "high_bill_flag",
    when(col("estimated_total_bill") > 1500, "Yes")
    .otherwise("No")
)

# 9. Rename patient_name to full_name
patients_df = patients_df.withColumnRenamed("patient_name", "full_name")

# 10. Drop temporary column
#full_df = full_df.drop("test_cost")
```

##Part 3 — Joins

```python
from pyspark.sql.functions import col, count

# 1. Join patients with visits
patient_visits_df = visits_df.join(
    patients_df,
    on="patient_id",
    how="inner"
)
patient_visits_df.show()

# 2. Join visits with doctors
visits_doctors_df = visits_df.join(
    doctors_df,
    on="doctor_id",
    how="inner"
)
visits_doctors_df.show()

# 3. Join visits with payments
visits_payments_df = visits_df.join(
    payments_df,
    on="visit_id",
    how="inner"
)
visits_payments_df.show()

# 4. Create final joined DataFrame (patients + visits + doctors + payments)
final_df = visits_df \
    .join(patients_df, "patient_id", "inner") \
    .join(doctors_df, "doctor_id", "inner") \
    .join(payments_df, "visit_id", "inner")
final_df.show()

# 5. Show required columns
final_df.select(
    "full_name",         
    "city",
    "doctor_name",
    "specialization",
    "visit_status",
    "bill_amount"
).show()

# 6. Find visits where patient city and doctor city are different
final_df.filter(
    col("city") != col("doctor_city")   # make sure doctor_city column exists
).show()

# 7. Find completed visits with paid payments
final_df.filter(
    (col("visit_status") == "Completed") &
    (col("payment_status") == "Paid")
).show()

# 8. Find pending visits with pending payments
final_df.filter(
    (col("visit_status") == "Pending") &
    (col("payment_status") == "Pending")
).show()

# 9. Find cancelled visits with cancelled payments
final_df.filter(
    (col("visit_status") == "Cancelled") &
    (col("payment_status") == "Cancelled")
).show()

# 10. Find patients who visited more than once
patients_multiple_visits = visits_df.groupBy("patient_id") \
    .agg(count("*").alias("visit_count")) \
    .filter(col("visit_count") > 1)
patients_multiple_visits.show()
```

##Part 4 — Aggregations

```python
from pyspark.sql.functions import col, count, sum, avg

# 1. Count patients by city
patients_df.groupBy("city") \
    .agg(count("*").alias("patient_count")) \
    .show()

# 2. Count patients by membership
patients_df.groupBy("membership") \
    .agg(count("*").alias("patient_count")) \
    .show()

# 3. Count doctors by specialization
doctors_df.groupBy("specialization") \
    .agg(count("*").alias("doctor_count")) \
    .show()

# 4. Count visits by status
visits_df.groupBy("visit_status") \
    .agg(count("*").alias("visit_count")) \
    .show()

# 5. Count payments by payment mode
payments_df.groupBy("payment_mode") \
    .agg(count("*").alias("payment_count")) \
    .show()

# 6. Calculate total bill amount
final_df.agg(
    sum("bill_amount").alias("total_bill_amount")
).show()

# 7. Calculate average bill amount
final_df.agg(
    avg("bill_amount").alias("avg_bill_amount")
).show()

# 8. Calculate total revenue by city
final_df.groupBy("city") \
    .agg(sum("bill_amount").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc()) \
    .show()

# 9. Calculate total revenue by specialization
final_df.groupBy("specialization") \
    .agg(sum("bill_amount").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc()) \
    .show()

# 10. Calculate total revenue by doctor
final_df.groupBy("doctor_name") \
    .agg(sum("bill_amount").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc()) \
    .show()
```

##Part 5 — Spark SQL

```python
# 1. Create temp views for all DataFrames
patients_df.createOrReplaceTempView("patients")
doctors_df.createOrReplaceTempView("doctors")
visits_df.createOrReplaceTempView("visits")
payments_df.createOrReplaceTempView("payments")
final_df.createOrReplaceTempView("final")

# 2. Show all patients using SQL
spark.sql("""
SELECT * FROM patients
""").show()

# 3. Find Cardiology visits using SQL
spark.sql("""
SELECT v.*
FROM visits v
JOIN doctors d ON v.doctor_id = d.doctor_id
WHERE d.specialization = 'Cardiology'
""").show()

# 4. Find revenue by city using SQL
spark.sql("""
SELECT city, SUM(bill_amount) AS total_revenue
FROM final
GROUP BY city
ORDER BY total_revenue DESC
""").show()

# 5. Find revenue by specialization using SQL
spark.sql("""
SELECT specialization, SUM(bill_amount) AS total_revenue
FROM final
GROUP BY specialization
ORDER BY total_revenue DESC
""").show()

# 6. Find top 5 highest bill visits
spark.sql("""
SELECT visit_id, bill_amount
FROM final
ORDER BY bill_amount DESC
LIMIT 5
""").show()

# 7. Count visits per doctor
spark.sql("""
SELECT doctor_name, COUNT(*) AS visit_count
FROM final
GROUP BY doctor_name
ORDER BY visit_count DESC
""").show()

# 8. Count visits per patient
spark.sql("""
SELECT full_name, COUNT(*) AS visit_count
FROM final
GROUP BY full_name
ORDER BY visit_count DESC
""").show()

# 9. Find average bill by payment mode
spark.sql("""
SELECT payment_mode, AVG(bill_amount) AS avg_bill
FROM final
GROUP BY payment_mode
""").show()

# 10. Find patients with total billing above 10000
spark.sql("""
SELECT full_name, SUM(bill_amount) AS total_bill
FROM final
GROUP BY full_name
HAVING SUM(bill_amount) > 10000
ORDER BY total_bill DESC
""").show()
```

##Part 6 — Window Functions

```python
from pyspark.sql.functions import col, sum, row_number, dense_rank, rank
from pyspark.sql.window import Window

# 1. Rank patients by total bill within city
patient_city_window = Window.partitionBy("city").orderBy(col("total_bill").desc())

patient_total_df = final_df.groupBy("city", "full_name") \
    .agg(sum("bill_amount").alias("total_bill"))

patient_total_df = patient_total_df.withColumn(
    "rank_in_city",
    rank().over(patient_city_window)
)

patient_total_df.show()


# 2. Rank doctors by total revenue within specialization
doctor_spec_window = Window.partitionBy("specialization").orderBy(col("total_revenue").desc())

doctor_total_df = final_df.groupBy("specialization", "doctor_name") \
.agg(sum("bill_amount").alias("total_revenue"))

doctor_total_df = doctor_total_df.withColumn(
    "rank_in_specialization",
    rank().over(doctor_spec_window)
)

doctor_total_df.show()


# 3. Use ROW_NUMBER to find top billing patient per city
rownum_window = Window.partitionBy("city").orderBy(col("total_bill").desc())

top_patient_per_city = patient_total_df.withColumn(
    "row_num",
    row_number().over(rownum_window)
).filter(col("row_num") == 1)

top_patient_per_city.show()


# 4. Use DENSE_RANK to rank doctors by consultation fee within specialization
fee_window = Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())

doctor_fee_rank = doctors_df.withColumn(
    "dense_rank_fee",
    dense_rank().over(fee_window)
)

doctor_fee_rank.show()


# 5. Find top 2 doctors by revenue
top_doctors = doctor_total_df.orderBy(col("total_revenue").desc()).limit(2)
top_doctors.show()


# 6. Create running total revenue by visit date
date_window = Window.orderBy("visit_date").rowsBetween(Window.unboundedPreceding, 0)

running_revenue_date = final_df.withColumn(
    "running_total_revenue",
    sum("bill_amount").over(date_window)
)

running_revenue_date.select("visit_date", "bill_amount", "running_total_revenue").show()


# 7. Create running total revenue by doctor
doctor_running_window = Window.partitionBy("doctor_name") \
    .orderBy("visit_date") \
    .rowsBetween(Window.unboundedPreceding, 0)

running_revenue_doctor = final_df.withColumn(
    "running_total_by_doctor",
    sum("bill_amount").over(doctor_running_window)
)

running_revenue_doctor.select("doctor_name", "visit_date", "bill_amount", "running_total_by_doctor").show()


# 8. Rank cities by total revenue
city_revenue_df = final_df.groupBy("city") \
    .agg(sum("bill_amount").alias("total_revenue"))

city_rank_window = Window.orderBy(col("total_revenue").desc())

city_revenue_df = city_revenue_df.withColumn(
    "city_rank",
    rank().over(city_rank_window)
)

city_revenue_df.show()


# 9. Rank specializations by total revenue
spec_revenue_df = final_df.groupBy("specialization") \
    .agg(sum("bill_amount").alias("total_revenue"))

spec_rank_window = Window.orderBy(col("total_revenue").desc())

spec_revenue_df = spec_revenue_df.withColumn(
    "spec_rank",
    rank().over(spec_rank_window)
)

spec_revenue_df.show()


# 10. Find highest bill visit per payment mode
payment_window = Window.partitionBy("payment_mode").orderBy(col("bill_amount").desc())

highest_bill_per_mode = final_df.withColumn(
    "row_num",
    row_number().over(payment_window)
).filter(col("row_num") == 1)

highest_bill_per_mode.show()
```

##Part 7 — Delta Lake Core

```python
from pyspark.sql.functions import col
from delta.tables import DeltaTable

# 0. Set catalog & schema
spark.sql("USE CATALOG my_catalog")
spark.sql("USE SCHEMA my_schema")

# 1. Drop table if exists (fresh start)
spark.sql("DROP TABLE IF EXISTS hospital_delta")

# 2. Rename amount → bill_amount (IMPORTANT)
final_df = final_df.withColumnRenamed("amount", "bill_amount")

# 3. Save final data as Delta table
final_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("hospital_delta")

# 4. Insert 2 new records
spark.sql("""
INSERT INTO hospital_delta
SELECT * FROM hospital_delta LIMIT 2
""")

# 5. Update visit status (Pending → Completed)
spark.sql("""
UPDATE hospital_delta
SET visit_status = 'Completed'
WHERE visit_status = 'Pending'
""")

# 6. Update payment status (Pending → Paid)
spark.sql("""
UPDATE hospital_delta
SET payment_status = 'Paid'
WHERE payment_status = 'Pending'
""")

# 7. Delete cancelled visits
spark.sql("""
DELETE FROM hospital_delta
WHERE visit_status = 'Cancelled'
""")

# 8. Create separate table for completed visits
spark.sql("DROP TABLE IF EXISTS completed_visits")

spark.sql("""
CREATE TABLE completed_visits AS
SELECT * FROM hospital_delta
WHERE visit_status = 'Completed'
""")

# 9. View history
spark.sql("DESCRIBE HISTORY hospital_delta").show()

# 10. Time travel (version 0)
old_df = spark.read.format('delta').option('versionAsOf', 0).table('hospital_delta')
old_df.show()

# Compare
latest_df = spark.table("hospital_delta")
print("Latest:", latest_df.count())
print("Old:", old_df.count())

# 11. Vacuum dry run
spark.sql("""
VACUUM hospital_delta RETAIN 168 HOURS DRY RUN
""").show()
```

##Part 8 — Merge / Upsert

```python
daily_visits_data = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1),
(11,1011,205,"2024-03-06","Completed",3),
(20,1020,206,"2024-03-10","Completed",2),
(28,1018,203,"2024-03-15","Pending",2)
]
 
daily_visits_columns = [
"visit_id","patient_id","doctor_id","visit_date","visit_status","tests_count"
]
 
daily_visits_df = spark.createDataFrame(daily_visits_data, daily_visits_columns)
```

```python
from pyspark.sql.functions import col

# 0. Clean start
spark.sql("DROP TABLE IF EXISTS visits_delta")

# 1. Create initial visits table (auto schema)
visits_df_clean = visits_df.select(
    "visit_id","patient_id","doctor_id","visit_date","visit_status"
).withColumn("tests_count", col("visit_id") * 0)

visits_df_clean.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("visits_delta")

# 2. Daily updates (your given data)
daily_visits_data = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1),
(11,1011,205,"2024-03-06","Completed",3),
(20,1020,206,"2024-03-10","Completed",2),
(28,1018,203,"2024-03-15","Pending",2)
]

columns = ["visit_id","patient_id","doctor_id","visit_date","visit_status","tests_count"]

daily_visits_df = spark.createDataFrame(daily_visits_data, columns)

# 3. Create temp view
daily_visits_df.createOrReplaceTempView("daily_updates")

# 4. MERGE (UPSERT)
spark.sql("""
MERGE INTO visits_delta AS target
USING daily_updates AS source
ON target.visit_id = source.visit_id

WHEN MATCHED THEN UPDATE SET
  target.patient_id = source.patient_id,
  target.doctor_id = source.doctor_id,
  target.visit_date = source.visit_date,
  target.visit_status = source.visit_status,
  target.tests_count = source.tests_count

WHEN NOT MATCHED THEN INSERT *
""")

# 5. Verify updated records
spark.sql("""
SELECT * FROM visits_delta WHERE visit_id IN (11,20)
""").show()

# 6. Verify inserted records
spark.sql("""
SELECT * FROM visits_delta WHERE visit_id IN (26,27,28)
""").show()

# 7. Full table check
spark.sql("SELECT * FROM visits_delta ORDER BY visit_id").show()

# 8. History check
spark.sql("DESCRIBE HISTORY visits_delta").show()
```

##Part 9 — Parquet to Delta

```python
from pyspark.sql.functions import col

# 1. Save patients data as Parquet
patients_df.write \
    .format("parquet") \
    .mode("overwrite") \
    .format("delta") .mode("overwrite") .saveAsTable("patients_parquet")


# 2. Read the Parquet data
parquet_df = spark.read.table("patients_parquet")
parquet_df.show()


# 3. Convert the Parquet data into Delta table format
parquet_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("patients_delta")


# 4. Validate the Delta table
spark.sql("SELECT * FROM patients_delta").show()


# 5. Compare Parquet and Delta behavior
print("Parquet count:", spark.table("patients_parquet").count())
print("Delta count:", spark.table("patients_delta").count())


# 6. Perform update on Delta table
spark.sql("""
UPDATE patients_delta
SET age = age + 1
WHERE age IS NOT NULL
""")
spark.sql("SELECT * FROM patients_delta").show()


# 7. Try update on Parquet (will fail or not supported)
try:
    spark.sql("""
    UPDATE patients_parquet
    SET age = age + 1
    """)
except Exception as e:
    print("Parquet update failed :", e)


# 8. Why is update easier in Delta than Parquet?
# Delta supports ACID transactions, so it can modify data safely.
# Parquet is immutable → requires full rewrite.
```
