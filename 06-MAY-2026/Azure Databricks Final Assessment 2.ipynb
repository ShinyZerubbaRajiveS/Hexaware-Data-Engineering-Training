##Part 2 — DataFrame Transformations

```python
from pyspark.sql.functions import col, to_date, when, upper

#  1: Convert order_date into date type
orders_df = orders_df.withColumn(
    "order_date",
    to_date(col("order_date"), "yyyy-MM-dd")
)


#  2: Add total_order_value column (quantity * price)
orders_df = orders_df.join(
    products_df.select("product_id", "price"),
    on="product_id",
    how="left"
).withColumn(
    "total_order_value",
    col("quantity") * col("price")
)


#  3: Create stock_status column
products_df = products_df.withColumn(
    "stock_status",
    when(col("stock_quantity") < 20, "Low")
    .when(col("stock_quantity") < 50, "Medium")
    .otherwise("High")
)


#  4: Create order_priority column
orders_df = orders_df.withColumn(
    "order_priority",
    when(col("quantity") >= 50, "High")
    .when(col("quantity") >= 20, "Medium")
    .otherwise("Low")
)


#  5: Create expensive_product_flag column
products_df = products_df.withColumn(
    "expensive_product_flag",
    when(col("price") > 20000, "Yes").otherwise("No")
)


#  6: Convert category names into uppercase
products_df = products_df.withColumn(
    "category",
    upper(col("category"))
)


#  7: Rename warehouse_city to inventory_city
products_df = products_df.withColumnRenamed(
    "warehouse_city",
    "inventory_city"
)


#  8: Create inventory_value column (price * stock_quantity)
products_df = products_df.withColumn(
    "inventory_value",
    col("price") * col("stock_quantity")
)


#  9: Drop temporary columns (price added during join if not needed)
orders_df = orders_df.drop("price")


#  10: Create low_stock_flag column
products_df = products_df.withColumn(
    "low_stock_flag",
    when(col("stock_quantity") < 20, "Yes").otherwise("No")
)
```

##Part 3 — Joins

```python
from pyspark.sql.functions import col, count

#  1: Join products with orders
products_orders_df = orders_df.join(
    products_df,
    on="product_id",
    how="inner"
)


#  2: Join orders with suppliers
orders_suppliers_df = orders_df.join(
    suppliers_df,
    on="supplier_id",
    how="inner"
)


#  3: Join orders with payments
orders_payments_df = orders_df.join(
    payments_df,
    on="order_id",
    how="inner"
)


#  4: Create final joined DataFrame (orders + products + suppliers + payments)
final_df = orders_df \
    .join(products_df, on="product_id", how="inner") \
    .join(suppliers_df, on="supplier_id", how="inner") \
    .join(payments_df, on="order_id", how="inner")


#  5: Display product name, supplier name, quantity, bill amount
final_df.select(
    "product_name",
    "supplier_name",
    "quantity",
    "bill_amount"
).show()


#  6: Find suppliers serving different cities (supplier_city != inventory_city)
suppliers_diff_city_df = final_df.filter(
    col("supplier_city") != col("inventory_city")
).select(
    "supplier_name", "supplier_city", "inventory_city"
).distinct()

suppliers_diff_city_df.show()


#  7: Find delivered orders with paid payments
delivered_paid_df = final_df.filter(
    (col("order_status") == "Delivered") &
    (col("payment_status") == "Paid")
)

delivered_paid_df.show()


#  8: Find pending orders with pending payments
pending_pending_df = final_df.filter(
    (col("order_status") == "Pending") &
    (col("payment_status") == "Pending")
)

pending_pending_df.show()


#  9: Find cancelled orders with cancelled payments
cancelled_cancelled_df = final_df.filter(
    (col("order_status") == "Cancelled") &
    (col("payment_status") == "Cancelled")
)

cancelled_cancelled_df.show()


#  10: Find products ordered multiple times
products_multiple_orders_df = final_df.groupBy(
    "product_id", "product_name"
).agg(
    count("order_id").alias("order_count")
).filter(
    col("order_count") > 1
)

products_multiple_orders_df.show()
```

##Part 4 — Aggregations

```python
from pyspark.sql.functions import col, count, sum, avg, desc, asc

#   1: Count products by category
products_by_category_df = products_df.groupBy(
    "category"
).agg(
    count("product_id").alias("product_count")
)

products_by_category_df.show()


#   2: Count orders by status
orders_by_status_df = orders_df.groupBy(
    "order_status"
).agg(
    count("order_id").alias("order_count")
)

orders_by_status_df.show()


#   3: Count suppliers by city
suppliers_by_city_df = suppliers_df.groupBy(
    "supplier_city"
).agg(
    count("supplier_id").alias("supplier_count")
)

suppliers_by_city_df.show()


#   4: Calculate total revenue
total_revenue_df = payments_df.agg(
    sum("bill_amount").alias("total_revenue")
)

total_revenue_df.show()


#   5: Calculate average bill amount
avg_bill_df = payments_df.agg(
    avg("bill_amount").alias("average_bill_amount")
)

avg_bill_df.show()


#   6: Calculate total revenue by category
revenue_by_category_df = final_df.groupBy(
    "category"
).agg(
    sum("bill_amount").alias("total_revenue")
)

revenue_by_category_df.show()


#   7: Calculate total revenue by supplier
revenue_by_supplier_df = final_df.groupBy(
    "supplier_name"
).agg(
    sum("bill_amount").alias("total_revenue")
)

revenue_by_supplier_df.show()


#   8: Calculate total revenue by city (inventory city)
revenue_by_city_df = final_df.groupBy(
    "inventory_city"
).agg(
    sum("bill_amount").alias("total_revenue")
)

revenue_by_city_df.show()


#   9: Find top-selling products (by quantity)
top_selling_products_df = final_df.groupBy(
    "product_id", "product_name"
).agg(
    sum("quantity").alias("total_quantity_sold")
).orderBy(
    desc("total_quantity_sold")
)

top_selling_products_df.show()


#   10: Find lowest-selling products (by quantity)
lowest_selling_products_df = final_df.groupBy(
    "product_id", "product_name"
).agg(
    sum("quantity").alias("total_quantity_sold")
).orderBy(
    asc("total_quantity_sold")
)

lowest_selling_products_df.show()
```

##Part 5 — Spark SQL

```python
#   1: Create temp views from DataFrames (PySpark way - REQUIRED)
products_df.createOrReplaceTempView("products_view")
suppliers_df.createOrReplaceTempView("suppliers_view")
orders_df.createOrReplaceTempView("orders_view")
payments_df.createOrReplaceTempView("payments_view")
final_df.createOrReplaceTempView("final_view")


#   2: Show all products using SQL
spark.sql("""
SELECT * FROM products_view
""").show()


#   3: Find electronics orders using SQL
spark.sql("""
SELECT o.*, p.product_name, p.category
FROM orders_view o
JOIN products_view p
ON o.product_id = p.product_id
WHERE p.category = 'ELECTRONICS'
""").show()


#   4: Find revenue by category
spark.sql("""
SELECT category, SUM(bill_amount) AS total_revenue
FROM final_view
GROUP BY category
""").show()


#   5: Find revenue by supplier
spark.sql("""
SELECT supplier_name, SUM(bill_amount) AS total_revenue
FROM final_view
GROUP BY supplier_name
""").show()


#   6: Find top 5 highest bill orders
spark.sql("""
SELECT order_id, bill_amount
FROM payments_view
ORDER BY bill_amount DESC
LIMIT 5
""").show()


#   7: Count orders per supplier
spark.sql("""
SELECT s.supplier_name, COUNT(o.order_id) AS order_count
FROM orders_view o
JOIN suppliers_view s
ON o.supplier_id = s.supplier_id
GROUP BY s.supplier_name
""").show()


#   8: Count orders per category
spark.sql("""
SELECT p.category, COUNT(o.order_id) AS order_count
FROM orders_view o
JOIN products_view p
ON o.product_id = p.product_id
GROUP BY p.category
""").show()


#   9: Find average payment per mode
spark.sql("""
SELECT payment_mode, AVG(bill_amount) AS avg_payment
FROM payments_view
GROUP BY payment_mode
""").show()


#   10: Find products generating revenue above 100000
spark.sql("""
SELECT product_id, product_name, SUM(bill_amount) AS total_revenue
FROM final_view
GROUP BY product_id, product_name
HAVING SUM(bill_amount) > 100000
""").show()
```

##Part 6 — Window Functions

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, sum, rank, dense_rank, row_number, desc, to_date

#   1: Rank products by revenue within category
product_revenue_df = final_df.groupBy(
    "category", "product_id", "product_name"
).agg(
    sum("bill_amount").alias("product_revenue")
)

w_category = Window.partitionBy("category").orderBy(desc("product_revenue"))

product_rank_df = product_revenue_df.withColumn(
    "product_rank",
    rank().over(w_category)
)

product_rank_df.show()


#   2: Rank suppliers by revenue within city
supplier_revenue_city_df = final_df.groupBy(
    "inventory_city", "supplier_id", "supplier_name"
).agg(
    sum("bill_amount").alias("supplier_revenue")
)

w_city = Window.partitionBy("inventory_city").orderBy(desc("supplier_revenue"))

supplier_rank_city_df = supplier_revenue_city_df.withColumn(
    "supplier_rank",
    rank().over(w_city)
)

supplier_rank_city_df.show()


#   3: Use ROW_NUMBER to find top product per category
top_product_per_category_df = product_revenue_df.withColumn(
    "row_num",
    row_number().over(w_category)
).filter(
    col("row_num") == 1
)

top_product_per_category_df.show()


#   4: Use DENSE_RANK to rank suppliers by bill amount
w_supplier_bill = Window.partitionBy("supplier_id").orderBy(desc("bill_amount"))

supplier_dense_rank_df = final_df.withColumn(
    "dense_rank",
    dense_rank().over(w_supplier_bill)
)

supplier_dense_rank_df.show()


#   5: Find top 2 suppliers by revenue
supplier_total_revenue_df = final_df.groupBy(
    "supplier_id", "supplier_name"
).agg(
    sum("bill_amount").alias("total_revenue")
)

w_supplier = Window.orderBy(desc("total_revenue"))

top_suppliers_df = supplier_total_revenue_df.withColumn(
    "rank",
    rank().over(w_supplier)
).filter(
    col("rank") <= 2
)

top_suppliers_df.show()


#   6: Create running total revenue by order date
final_with_date_df = final_df.withColumn(
    "order_date",
    to_date(col("order_date"))
)

w_date = Window.orderBy("order_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)

running_revenue_date_df = final_with_date_df.withColumn(
    "running_total_revenue",
    sum("bill_amount").over(w_date)
)

running_revenue_date_df.select("order_date", "bill_amount", "running_total_revenue").show()


#   7: Create running total revenue by supplier
w_supplier_running = Window.partitionBy("supplier_id").orderBy("order_date") \
    .rowsBetween(Window.unboundedPreceding, Window.currentRow)

running_revenue_supplier_df = final_with_date_df.withColumn(
    "running_supplier_revenue",
    sum("bill_amount").over(w_supplier_running)
)

running_revenue_supplier_df.select(
    "supplier_name", "order_date", "bill_amount", "running_supplier_revenue"
).show()


#   8: Rank cities by revenue
city_revenue_df = final_df.groupBy(
    "inventory_city"
).agg(
    sum("bill_amount").alias("city_revenue")
)

w_city_rank = Window.orderBy(desc("city_revenue"))

city_rank_df = city_revenue_df.withColumn(
    "city_rank",
    rank().over(w_city_rank)
)

city_rank_df.show()


#   9: Rank categories by revenue
category_revenue_df = final_df.groupBy(
    "category"
).agg(
    sum("bill_amount").alias("category_revenue")
)

w_category_rank = Window.orderBy(desc("category_revenue"))

category_rank_df = category_revenue_df.withColumn(
    "category_rank",
    rank().over(w_category_rank)
)

category_rank_df.show()


#   10: Find highest bill order per payment mode
w_payment_mode = Window.partitionBy("payment_mode").orderBy(desc("bill_amount"))

highest_bill_per_mode_df = final_df.withColumn(
    "row_num",
    row_number().over(w_payment_mode)
).filter(
    col("row_num") == 1
)

highest_bill_per_mode_df.select(
    "payment_mode", "order_id", "bill_amount"
).show()
```

##Part 7 — Delta Lake Core

```python
from pyspark.sql.functions import col
from delta.tables import DeltaTable

#   1: Save final joined data as Delta table (Unity Catalog - using my_catalog.my_schema)
final_df.write.format("delta").mode("overwrite").saveAsTable("my_catalog.my_schema.final_orders_delta")


#   2: Insert new order records
spark.sql("""
INSERT INTO my_catalog.my_schema.final_orders_delta
SELECT * FROM final_view WHERE order_id = 301
""")


#   3: Update one pending order
spark.sql("""
UPDATE my_catalog.my_schema.final_orders_delta
SET order_status = 'Delivered'
WHERE order_status = 'Pending'
AND order_id = 304
""")


#   4: Update one pending payment
spark.sql("""
UPDATE my_catalog.my_schema.final_orders_delta
SET payment_status = 'Paid'
WHERE payment_status = 'Pending'
AND order_id = 304
""")


#   5: Delete cancelled orders
spark.sql("""
DELETE FROM my_catalog.my_schema.final_orders_delta
WHERE order_status = 'Cancelled'
""")


#   6: Create clean_orders Delta table
spark.sql("""
CREATE OR REPLACE TABLE my_catalog.my_schema.clean_orders_delta AS
SELECT *
FROM my_catalog.my_schema.final_orders_delta
WHERE order_status = 'Delivered'
AND payment_status = 'Paid'
""")


#   7: Query Delta history
spark.sql("""
DESCRIBE HISTORY my_catalog.my_schema.final_orders_delta
""").show(truncate=False)


#   8: Query previous version using Time Travel
spark.sql("""
SELECT *
FROM my_catalog.my_schema.final_orders_delta VERSION AS OF 0
""").show()


#   9: Compare old and latest versions
old_df = spark.sql("SELECT * FROM my_catalog.my_schema.final_orders_delta VERSION AS OF 0")
latest_df = spark.sql("SELECT * FROM my_catalog.my_schema.final_orders_delta")

difference_df = latest_df.subtract(old_df)
difference_df.show()


#   10: Run VACUUM DRY RUN
spark.sql("""
VACUUM my_catalog.my_schema.final_orders_delta RETAIN 168 HOURS DRY RUN
""").show()
```

```python
from pyspark.sql.functions import col, to_date

#   1: Create Delta target table (only required columns to avoid schema mismatch)
orders_df.select(
    "order_id", "product_id", "supplier_id",
    "order_date", "quantity", "order_status"
).write.format("delta").mode("overwrite") \
 .saveAsTable("my_catalog.my_schema.orders_delta_target")


#   2: Load initial orders (verify data)
spark.sql("""
SELECT * FROM my_catalog.my_schema.orders_delta_target
""").show()


#   3: Create temp view for updates
daily_orders_df.createOrReplaceTempView("daily_orders_view")


#   4: Merge updates into target table
spark.sql("""
MERGE INTO my_catalog.my_schema.orders_delta_target AS target
USING daily_orders_view AS source
ON target.order_id = source.order_id

WHEN MATCHED THEN UPDATE SET
  target.product_id = source.product_id,
  target.supplier_id = source.supplier_id,
  target.order_date = source.order_date,
  target.quantity = source.quantity,
  target.order_status = source.order_status

WHEN NOT MATCHED THEN INSERT (
  order_id,
  product_id,
  supplier_id,
  order_date,
  quantity,
  order_status
)
VALUES (
  source.order_id,
  source.product_id,
  source.supplier_id,
  source.order_date,
  source.quantity,
  source.order_status
)
""")


#   5: Update existing records (verification)
spark.sql("""
SELECT * FROM my_catalog.my_schema.orders_delta_target
WHERE order_id IN (304, 319)
""").show()


#   6: Insert new records (verification)
spark.sql("""
SELECT * FROM my_catalog.my_schema.orders_delta_target
WHERE order_id IN (321, 322, 323)
""").show()


#   7: Verify updated order IDs
spark.sql("""
SELECT order_id, order_status
FROM my_catalog.my_schema.orders_delta_target
WHERE order_id IN (304, 319)
""").show()


#   8: Verify inserted order IDs
spark.sql("""
SELECT order_id, order_status
FROM my_catalog.my_schema.orders_delta_target
WHERE order_id IN (321, 322, 323)
""").show()


#   9: Check Delta history after merge
spark.sql("""
DESCRIBE HISTORY my_catalog.my_schema.orders_delta_target
""").show(truncate=False)


#   10: Explain why MERGE is important
# MERGE (Upsert) is used to handle incremental data loads.
# It updates existing records and inserts new records in one operation.
# Prevents duplicates and maintains data consistency.
# Essential for real-world pipelines like daily/streaming data ingestion.
```

##Part 9 — Parquet to Delta

```python
from pyspark.sql.functions import col


# 1. Save products data as Parquet
products_df.write \
    .format("parquet") \
    .mode("overwrite") \
    .format("delta") .mode("overwrite") .saveAsTable("products_parquet")


# 2. Read Parquet data
parquet_df = spark.table("products_parquet")
parquet_df.show()


# 3. Convert Parquet into Delta
parquet_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("products_delta")


# 4. Validate Delta table
spark.sql("""
SELECT * FROM products_delta
""").show()


# 5. Compare Parquet vs Delta behavior
# Parquet → file format only, no ACID, no update/delete
# Delta → ACID transactions, supports update/delete/merge, time travel
print("Parquet count:", parquet_df.count())
print("Delta count:", spark.table("products_delta").count())


# 6. Perform update on Delta table
spark.sql("""
UPDATE products_delta
SET price = price + 500
WHERE category = 'ELECTRONICS'
""")


# 7. Verify update
spark.sql("""
SELECT product_name, category, price
FROM products_delta
WHERE category = 'ELECTRONICS'
""").show()


# 8. Why Delta supports updates better than Parquet?
# Delta Lake maintains a transaction log (_delta_log)
# Enables ACID transactions
# Only modified data files are rewritten (not full dataset)
# Supports versioning (time travel) and concurrent writes safely
```
