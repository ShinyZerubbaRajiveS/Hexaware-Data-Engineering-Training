## **Final Capstone Mini Project**


1. Join all tables
2. Calculate revenue

3. Find top customers
4. Find top products
5. Find revenue per city
6. Rank customers by revenue
7. Save final dataset

```python
# 1
from pyspark.sql.functions import col

full_df = orders \
    .join(customers, "customer_id") \
    .join(order_items, "order_id") \
    .join(products, "product_id")
full_df.show()
```

```python
# 2
full_df = full_df.withColumn(
    "revenue",
    col("quantity") * col("product_price")
)
full_df.show()
```

```python
# 3
top_customers = full_df.groupBy("customer_name") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())
top_customers.show()
```

```python
# 4
top_products = full_df.groupBy("product_name") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())
top_products.show()
```

```python
# 5
revenue_per_city = full_df.groupBy("city") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())
revenue_per_city.show()
```

```python
# 6
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
window_spec = Window.orderBy(col("total_revenue").desc())
ranked_customers = top_customers.withColumn(
    "rank",
    rank().over(window_spec)
)
ranked_customers.show()
```

```python
# 7
final_df = full_df.select(
    "customer_name",
    "city",
    "product_name",
    "category",
    "quantity",
    "product_price",
    "revenue"
)
final_df.write.mode("overwrite").csv("ecommerce_report", header=True)
```
