```python

from pyspark.sql.functions import col

# 21. Sort employees by salary ascending
df.orderBy(col("salary").asc()).show()


# 22. Sort employees by salary descending
df.orderBy(col("salary").desc()).show()


# 23. Sort employees by experience ascending
df.orderBy(col("experience").asc()).show()


# 24. Sort employees by experience descending
df.orderBy(col("experience").desc()).show()


# 25. Sort employees by department
df.orderBy(col("department").asc()).show()


# 26. Sort employees by city
df.orderBy(col("city").asc()).show()


# 27. Sort employees by department and salary
df.orderBy(col("department").asc(), col("salary").desc()).show()


# 28. Show top 5 highest paid employees
df.orderBy(col("salary").desc()).limit(5).show()


# 29. Show bottom 5 lowest paid employees
df.orderBy(col("salary").asc()).limit(5).show()


# 30. Show top 3 most experienced employees
df.orderBy(col("experience").desc()).limit(3).show()
```

```python

from pyspark.sql.functions import col, lit

# 31. Create column annual_salary
df = df.withColumn("annual_salary", col("salary") * 12)
df.show()


# 32. Create column salary_in_thousands
df = df.withColumn("salary_in_thousands", col("salary") / 1000)
df.show()


# 33. Create column experience_double
df = df.withColumn("experience_double", col("experience") * 2)
df.show()


# 34. Create column company with value "GlobalTech"
df = df.withColumn("company", lit("GlobalTech"))
df.show()


# 35. Create column salary_after_bonus (salary + 5000)
df = df.withColumn("salary_after_bonus", col("salary") + 5000)
df.show()


# 36. Create column salary_after_tax (salary * 0.9)
df = df.withColumn("salary_after_tax", col("salary") * 0.9)
df.show()


# 37. Create column experience_next_year
df = df.withColumn("experience_next_year", col("experience") + 1)
df.show()


# 38. Create column salary_per_year_of_experience
df = df.withColumn(
    "salary_per_year_of_experience",
    col("salary") / col("experience")
)
df.show()


# 39. Rename column salary to monthly_salary
df = df.withColumnRenamed("salary", "monthly_salary")
df.show()


# 40. Drop column experience_double
df = df.drop("experience_double")
df.show()
```

```python
from pyspark.sql.functions import col, avg
from pyspark.sql.functions import sum as _sum


# 61. Remove duplicate rows (if any)
df = df.dropDuplicates()
df.show()


# 62. Check if there are null values in the dataset
df.select([_sum(col(c).isNull().cast("int")).alias(c) for c in df.columns]).show()


# 63. Replace null salary with 0
df = df.fillna({"monthly_salary": 0})
df.show()


# 64. Drop rows where department is null
df = df.dropna(subset=["department"])
df.show()


# 65. Fill missing experience with average experience
avg_exp = df.select(avg("experience")).collect()[0][0]
df = df.fillna({"experience": avg_exp})
df.show()


# 66. Check number of rows after cleaning
df.count()


# 67. Replace null department with "Unknown"
df = df.fillna({"department": "Unknown"})
df.show()


# 68. Remove rows with salary less than 60000
df = df.filter(col("monthly_salary") >= 60000)
df.show()


# 69. Remove employees with experience less than 2
df = df.filter(col("experience") >= 2)
df.show()


# 70. Display cleaned dataset
display(df)
```
