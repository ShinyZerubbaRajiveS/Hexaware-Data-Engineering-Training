```python
# 61
from pyspark.sql.functions import when
patients = patients.withColumn(
    "age_group",
    when(patients.age < 30, "Young")
    .when((patients.age >= 30) & (patients.age <= 40), "Adult")
    .otherwise("Senior")
)
patients.show()
```

```python
# 62
from pyspark.sql.functions import lit
patients = patients.withColumn("hospital_name", lit("BotCampus Hospital"))
patients.show()
```

```python
# 63
doctors = doctors.withColumn("fee_with_tax", doctors.consultation_fee * 1.18)
doctors.show()
```

```python
# 64
bills = bills.withColumn("bill_with_tax", bills.bill_amount * 1.18)
bills.show()
```

```python
# 65
bills = bills.withColumn("bill_in_thousands", bills.bill_amount / 1000)
bills.show()
```

```python
# 66
patients = patients.withColumn("country", lit("India"))
patients.show()
```

```python
# 67
from pyspark.sql.functions import concat, col, lit
doctors = doctors.withColumn(
    "doctor_label",
    concat(col("doctor_name"), lit(" - "), col("specialization"))
)
doctors.show()
```

```python
# 68
patients = patients.withColumn(
    "patient_label",
    concat(col("name"), lit(" - "), col("city"))
)
patients.show()
```

```python
# 69
bills = bills.withColumn(
    "high_bill_flag",
    when(bills.bill_amount > 1500, "Yes").otherwise("No")
)
bills.show()
```

## **Section 8 — When / Otherwise .**
71. Create patient age category: Young, Adult, Senior. .
72. Create bill category: High, Medium, Low. .
73. Create doctor fee category: Premium, Standard, Basic. .
74. Create appointment priority: Pending = High, Completed = Normal, Cancelled = Low. .
75. Create city zone: Hyderabad/Bangalore/Chennai = South, Mumbai/Delhi/Pune = Metro.

```python
# 71
from pyspark.sql.functions import when, col
patients = patients.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .when((col("age") >= 30) & (col("age") <= 40), "Adult")
    .otherwise("Senior")
)
patients.show()
```

```python
# 72
bills = bills.withColumn(
    "bill_category",
    when(col("bill_amount") > 1500, "High")
    .when((col("bill_amount") >= 1000) & (col("bill_amount") <= 1500), "Medium")
    .otherwise("Low")
)
bills.show()
```

```python
# 73
doctors = doctors.withColumn(
    "fee_category",
    when(col("consultation_fee") > 1500, "Premium")
    .when((col("consultation_fee") >= 1000) & (col("consultation_fee") <= 1500), "Standard")
    .otherwise("Basic")
)
doctors.show()
```

```python
# 74
appointments = appointments.withColumn(
    "priority",
    when(col("status") == "Pending", "High")
    .when(col("status") == "Completed", "Normal")
    .otherwise("Low")
)
appointments.show()
```

```python
# 75
patients = patients.withColumn(
    "city_zone",
    when(col("city").isin("Hyderabad", "Bangalore", "Chennai"), "South")
    .when(col("city").isin("Mumbai", "Delhi", "Pune"), "Metro")
)
patients.show()
```
