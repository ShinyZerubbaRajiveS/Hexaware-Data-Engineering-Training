```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Banking Project") \
    .getOrCreate()
```

```python
customers = spark.read.csv("bank_customers.csv", header=True, inferSchema=True)
accounts = spark.read.csv("accounts.csv", header=True, inferSchema=True)
transactions = spark.read.csv("transactions.csv", header=True, inferSchema=True)
branches = spark.read.csv("branches.csv", header=True, inferSchema=True)
profiles = spark.read.json("customer_profiles.json", multiLine=True)
```

## **Section 1 — DataFrame Basics**

```python
# 1. Show all customers
customers.show()

# 2. Show all accounts
accounts.show()

# 3. Show all transactions
transactions.show()

# 4. Show all branches
branches.show()

# 5. Print schema of customers
customers.printSchema()

# 6. Print schema of transactions
transactions.printSchema()

# 7. Count total customers
print("Total Customers:", customers.count())

# 8. Count total accounts
print("Total Accounts:", accounts.count())

# 9. Count total transactions
print("Total Transactions:", transactions.count())

# 10. Display first 5 transactions
transactions.show(5)
```

## **Section 2 — Select, Rename, Filter**

```python
from pyspark.sql.functions import col

# 11. Select customer name, city, and account type
customers.select("name", "city", "account_type").show()

# 12. Select account_id and balance
accounts.select("account_id", "balance").show()

# 13. Rename balance to current_balance
accounts.withColumnRenamed("balance", "current_balance").show()

# 14. Rename txn_type to transaction_type
transactions.withColumnRenamed("txn_type", "transaction_type").show()

# 15. Show customers from Hyderabad
customers.filter(col("city") == "Hyderabad").show()

# 16. Show customers older than 30
customers.filter(col("age") > 30).show()

# 17. Show Savings account customers
customers.filter(col("account_type") == "Savings").show()

# 18. Show accounts with balance greater than 100000
accounts.filter(col("balance") > 100000).show()

# 19. Show Credit transactions
transactions.filter(col("txn_type") == "Credit").show()

# 20. Show transactions with amount greater than 20000
transactions.filter(col("amount") > 20000).show()
```

## **Section 3 — Sorting and Limit**

```python
from pyspark.sql.functions import col

# 21. Sort customers by age ascending
customers.orderBy(col("age").asc()).show()

# 22. Sort customers by age descending
customers.orderBy(col("age").desc()).show()

# 23. Sort accounts by balance descending
accounts.orderBy(col("balance").desc()).show()

# 24. Show top 5 highest balance accounts
accounts.orderBy(col("balance").desc()).show(5)

# 25. Show bottom 5 lowest balance accounts
accounts.orderBy(col("balance").asc()).show(5)

# 26. Sort transactions by amount descending
transactions.orderBy(col("amount").desc()).show()

# 27. Show top 3 biggest transactions
transactions.orderBy(col("amount").desc()).show(3)

# 28. Sort transactions by txn_date
transactions.orderBy(col("txn_date")).show()

# 29. Sort customers by city and age
customers.orderBy(col("city"), col("age")).show()

# 30. Show latest 5 transactions
transactions.orderBy(col("txn_date").desc()).show(5)
```

## **Section 4 — Aggregations**

```python
from pyspark.sql.functions import sum, avg, max, min, count

# 31. Find total account balance
accounts.select(sum("balance")).show()

# 32. Find average account balance
accounts.select(avg("balance")).show()

# 33. Find highest account balance
accounts.select(max("balance")).show()

# 34. Find lowest account balance
accounts.select(min("balance")).show()

# 35. Count customers by city
customers.groupBy("city").count().show()

# 36. Count customers by account type
customers.groupBy("account_type").count().show()

# 37. Count transactions by transaction type
transactions.groupBy("txn_type").count().show()

# 38. Find total Credit transaction amount
transactions.filter(transactions.txn_type == "Credit") \
           .select(sum("amount")).show()

# 39. Find total Debit transaction amount
transactions.filter(transactions.txn_type == "Debit") \
           .select(sum("amount")).show()

# 40. Find average transaction amount
transactions.select(avg("amount")).show()
```

## **Section 5 — GroupBy Analytics**

```python
from pyspark.sql.functions import sum, avg, count

# 41 Find total balance by branch.
accounts.groupBy("branch").agg(sum("balance")).show()

# 42 Find average balance by branch.
accounts.groupBy("branch").agg(avg("balance")).show()

# 43 Count accounts by branch.
accounts.groupBy("branch").count().show()

# 44 Find total transaction amount by account_id.
transactions.groupBy("account_id").agg(sum("amount")).show()

# 45 Find transaction count by account_id.
transactions.groupBy("account_id").count().show()

# 46 Find total transaction amount by txn_type.
transactions.groupBy("txn_type").agg(sum("amount")).show()

# 47 Find transaction count by date.
transactions.groupBy("txn_date").count().show()

# 48 Find average customer age by city.
customers.groupBy("city").agg(avg("age")).show()

# 49 Find total balance by account type after joining customers and accounts.
customers.join(accounts, "customer_id") \
    .groupBy("account_type").agg(sum("balance")).show()

# 50 Find count of Savings and Current accounts by city.
customers.join(accounts, "customer_id") \
    .groupBy("city", "account_type").count().show()
```

## **Section 6 — Joins**

```python
# 51 Join customers with accounts using customer_id.
cust_acc = customers.join(accounts, "customer_id")
cust_acc.show()

# 52 Show customer name, city, branch, and balance.
cust_acc.select("name", "city", "branch", "balance").show()

# 53 Join accounts with transactions using account_id.
acc_txn = accounts.join(transactions, "account_id")
acc_txn.show()

# 54 Show account_id, txn_type, amount, and balance.
acc_txn.select("account_id", "txn_type", "amount", "balance").show()

# 55 Join accounts with branches using branch.
acc_branch = accounts.join(branches, "branch")
acc_branch.show()

# 56 Show branch, region, manager, and balance.
acc_branch.select("branch", "region", "manager", "balance").show()

# 57 Join customers, accounts, and transactions.
full_join = customers.join(accounts, "customer_id") \
                     .join(transactions, "account_id")

# 58 Show customer name, city, txn_type, amount, and txn_date.
full_join.select("name", "city", "txn_type", "amount", "txn_date").show()

# 59 Join customers, accounts, branches, and transactions.
full_all = customers.join(accounts, "customer_id") \
                    .join(branches, "branch") \
                    .join(transactions, "account_id")

# 60 Find total transaction amount by customer name.
full_all.groupBy("name").agg(sum("amount")).show()
```

## **Section 7 — Updating Data with withColumn**

```python
from pyspark.sql.functions import col, lit, concat

# 60 (continuation from joins section)
full_all.groupBy("name").sum("amount").show()

# 61. Add balance_in_lakhs
accounts.withColumn("balance_in_lakhs", col("balance") / 100000).show()

# 62. Add bank_name
accounts.withColumn("bank_name", lit("BotCampus Bank")).show()

# 63. Add annual_service_fee
accounts.withColumn("annual_service_fee", col("balance") * 0.01).show()

# 64. Add net_balance
accounts.withColumn("net_balance", col("balance") - (col("balance") * 0.01)).show()

# 65. Add is_high_balance
accounts.withColumn("is_high_balance", col("balance") > 100000).show()

# 66. Add txn_amount_in_k
transactions.withColumn("txn_amount_in_k", col("amount") / 1000).show()

# 67. Add country
customers.withColumn("country", lit("India")).show()

# 68. Add customer_label
customers.withColumn(
    "customer_label",
    concat(col("name"), lit(" - "), col("city"))
).show()

# 69. Add branch_label
branches.withColumn(
    "branch_label",
    concat(col("branch"), lit(" - "), col("region"))
).show()

# 70. Add risk_flag
transactions.withColumn("risk_flag", col("amount") > 40000).show()
```

## **Section 8 — when / otherwise**

```python
from pyspark.sql.functions import when

# 71 Create balance category: High if balance >= 100000, Medium if >= 50000, else Low.
accounts.withColumn("balance_category",
    when(accounts.balance >= 100000, "High")
    .when(accounts.balance >= 50000, "Medium")
    .otherwise("Low")
).show()

# 72 Create age group: Young if age < 30, Adult if age < 40, else Senior.
customers.withColumn("age_group",
    when(customers.age < 30, "Young")
    .when(customers.age < 40, "Adult")
    .otherwise("Senior")
).show()

# 73 Create transaction category: Large if amount >= 30000, Medium if >= 10000, else Small.
transactions.withColumn("txn_category",
    when(transactions.amount >= 30000, "Large")
    .when(transactions.amount >= 10000, "Medium")
    .otherwise("Small")
).show()

# 74 Create region priority: South = High Priority, North = Medium Priority, West = Watch.
branches.withColumn("priority",
    when(branches.region == "South", "High Priority")
    .when(branches.region == "North", "Medium Priority")
    .otherwise("Watch")
).show()

# 75 Create account type label: Savings = Retail, Current = Business.
customers.withColumn("account_label",
    when(customers.account_type == "Savings", "Retail")
    .otherwise("Business")
).show()
```

## **Section 9 — Date Functions**

```python
from pyspark.sql.functions import to_date, year, month, datediff, current_date

# 76 Convert signup_date to date type.
customers = customers.withColumn("signup_date", to_date("signup_date"))

# 77 Extract signup year.
customers.withColumn("signup_year", year("signup_date")).show()

# 78 Extract signup month.
customers.withColumn("signup_month", month("signup_date")).show()

# 79 Extract signup month.
transactions = transactions.withColumn("txn_date", to_date("txn_date"))

# 80 Extract transaction month.
transactions.withColumn("txn_month", month("txn_date")).show()

# 81 Count transactions by transaction date.
transactions.groupBy("txn_date").count().show()

# 82 Count transactions by month.
transactions.groupBy(month("txn_date")).count().show()

# 83 Find customers signed up after 2023-06-01
customers.filter(customers.signup_date > "2023-06-01").show()

# 84 Find days since signup
customers.withColumn("days_since_signup",
                     datediff(current_date(), customers.signup_date)).show()

# 85 Find days since transaction
transactions.withColumn("days_since_txn",
                        datediff(current_date(), transactions.txn_date)).show()
```

## **Section 10 — Window Functions**

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, rank, row_number, dense_rank, sum as _sum

# 86 Rank customers by balance within city
w = Window.partitionBy("city").orderBy(col("balance").desc())

customers.join(accounts, "customer_id") \
    .withColumn("rank", rank().over(w)).show()

# 87 Row number (top balance per city)
customers.join(accounts, "customer_id") \
    .withColumn("row_num", row_number().over(w)).show()

# 88 Dense rank transactions by txn_type
w2 = Window.partitionBy("txn_type").orderBy(col("amount").desc())

transactions.withColumn("dense_rank", dense_rank().over(w2)).show()

# 89 Top 2 transactions per account
w3 = Window.partitionBy("account_id").orderBy(col("amount").desc())

transactions.withColumn("rank", row_number().over(w3)) \
    .filter(col("rank") <= 2).show()

# 90 Running total per account
transactions.withColumn(
    "running_total",
    _sum("amount").over(
        Window.partitionBy("account_id").orderBy("txn_date")
    )
).show()

# 91 Rank branches by total balance
accounts.groupBy("branch") \
    .agg(_sum("balance").alias("total_balance")) \
    .withColumn(
        "rank",
        rank().over(Window.orderBy(col("total_balance").desc()))
    ).show()

# 92 Rank cities by total customer balance
customers.join(accounts, "customer_id") \
    .groupBy("city") \
    .agg(_sum("balance").alias("total_balance")) \
    .withColumn(
        "rank",
        rank().over(Window.orderBy(col("total_balance").desc()))
    ).show()

# 93 Rank customers by total transaction amount
full_all.groupBy("name") \
    .agg(_sum("amount").alias("total_amount")) \
    .withColumn(
        "rank",
        rank().over(Window.orderBy(col("total_amount").desc()))
    ).show()

# 94 Highest transaction customer per city
full_all.withColumn(
    "rank",
    rank().over(
        Window.partitionBy("city").orderBy(col("amount").desc())
    )
).filter(col("rank") == 1).show()

# 95 Highest balance customer per region
full_all.withColumn(
    "rank",
    rank().over(
        Window.partitionBy("region").orderBy(col("balance").desc())
    )
).filter(col("rank") == 1).show()
```
